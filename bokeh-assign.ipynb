In Bokeh, glyphs are visual shapes or markers that can be added to a plot to represent data points. Glyphs are used to create visualizations such as scatter plots, line plots, bar charts, and more. Bokeh provides a wide range of glyphs that can be customized to meet the needs of different visualizations.

Here's an example of how to add glyphs to a Bokeh plot:

```python
from bokeh.plotting import figure, output_file, show

# create some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# create a Bokeh figure
p = figure(title='Glyphs Example', x_axis_label='X-axis', y_axis_label='Y-axis')

# add a circle glyph to the plot
p.circle(x, y, size=20, color='navy', alpha=0.5)

# add a square glyph to the plot
p.square(x, y, size=15, color='red', alpha=0.8)

# specify the output file
output_file('glyphs_example.html')

# display the plot
show(p)
```

# question 3

In Bokeh, there are various ways to customize the appearance of a plot, including the axes, title, and legend. Here are some examples:

Customizing the axes:
To customize the appearance of the axes, we can use the axis method of the figure object. For example, to change the color of the x-axis labels to red, and make them bold, we can use the following code:
p.xaxis.axis_label_text_color = 'red'
p.xaxis.axis_label_text_font_style = 'bold'
We can similarly customize the appearance of the y-axis labels.

Customizing the title:
To customize the title of the plot, we can use the title attribute of the figure object. For example, to change the font size and color of the title to 18 and blue respectively, we can use the following code:
p.title.text = 'My Custom Title'
p.title.text_color = 'blue'
p.title.text_font_size = '18pt'

Customizing the legend:
To customize the legend, we can use the legend method of the figure object. For example, to change the position of the legend to the top right corner of the plot and make it horizontal, we can use the following code:
p.legend.location = 'top_right'
p.legend.orientation = 'horizontal'
